```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from apyori import apriori
```

```python
data = pd.read_csv("Market_Basket.csv", header=None)
```

```python
data
```

```python
data_list=[]
for i in range (len(data)):
    data_list.append([str(x) for x in data.values[i] if str(x) != "nan"])
data_list
```

```python
apr_list = list(apriori(data_list, min_support = 0.001, min_confidence = 0.3))
apr_list
```

```python
result = []
for i in apr_list:
    sub_list = []
    sub_list.append(i[0])
    sub_list.append(i[1])
    sub_list.append(i[2][0][2])
    sub_list.append(i[2][0][3])
    result.append(sub_list)
result_df = pd.DataFrame(result, columns=['Rules','Support','Confidence','Lift'])
result_df
```

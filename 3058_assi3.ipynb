```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import tree 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import ConfusionMatrixDisplay 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('Admission_Predict.csv')
data = data.drop(columns='Serial No.')
data
```

```python
y = [1 if each > 0.85 else 0 for each in data['Chance of Admit ']]
y
```

```python
x = data.iloc[:, :-1]
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
# Decision Tree Classifier using gini index
```

```python
model_dt = DecisionTreeClassifier().fit(x_train, y_train)
model_dt
```

```python
tree.plot_tree(model_dt, filled=False, feature_names=x_train.columns, class_names=["Not Admitted(0)", "Admitted(1)"], rounded=True)
plt.show()
```

```python
prediction_dt = model_dt.predict(x_test)
prediction_dt
```

```python
for i in range (len(prediction_dt)):
    pred = model_dt.predict(x_test.iloc[[i]])
    if (pred < 1):
        print("Student cannot be admitted.")
    else:
        print("Student can be admitted!!!")
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,prediction_dt)
plt.title('Decision Tree')
plt.show()
```

```python
#Report_lr
print(classification_report(y_test,prediction_dt))
```

```python
#Accuracy_lr
accuracy_score(y_test,prediction_dt)
```

```python
sample = x.iloc[[399]]
sample
```

```python
sample_prediction_dt = model_dt.predict(sample)
if (sample_prediction_dt < 1):
    print("Student cannot be admitted.")
else:
    print("Student can be admitted!!!")
```

```python
# Decision Tree Classifier using entropy
```

```python
model_dt_e = DecisionTreeClassifier(criterion='entropy').fit(x_train, y_train)
model_dt_e
```

```python
tree.plot_tree(model_dt_e, filled=True, feature_names=x_train.columns, class_names=["Not Admitted(0)", "Admitted(1)"], rounded=True)
plt.show()
```

```python
prediction_dt_e = model_dt_e.predict(x_test)
prediction_dt_e
```

```python
for i in range (len(prediction_dt_e)):
    pred_e = model_dt_e.predict(x_test.iloc[[i]])
    if (pred_e < 1):
        print("Student cannot be admitted.")
    else:
        print("Student can be admitted!!!")
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,prediction_dt_e)
plt.title('Decision Tree')
plt.show()
```

```python
#Report_dt_e
print(classification_report(y_test,prediction_dt_e))
```

```python
#Accuracy_dt_e
accuracy_score(y_test,prediction_dt_e)
```
